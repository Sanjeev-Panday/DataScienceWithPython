```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Random Normally distributed data
data = pd.Series(np.random.normal(loc=150,scale=50,size=10000))
# loc is Mean
# scale is standard deviation
```

```python
# seaborn distribution plot
plt.figure(figsize=(15,10))
# figsize tuble of width and height
sns.distplot(data)
plt.show()
```

```python
# 68% rule for normally distributed 
UC = data.mean() + data.std()
LC = data.mean() - data.std()

total = len(data)

# percentage of data between UC and LC
len(data.loc[(data > LC) & (data < UC)])/total * 100
```

```python
# Z transform
z_data = (data - data.mean()) / data.std()
```

```python
# After Z transform Mean = 0
round(z_data.mean())
```

```python
# After Z transform Std = 1
round(z_data.std())
```

```python
# Distribution plot z transformed data
sns.distplot(z_data)
plt.show()
```
