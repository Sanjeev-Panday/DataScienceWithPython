```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Random Normally distributed data
data = pd.Series(np.random.normal(loc=150,scale=50,size=10000))
# loc is Mean
# scale is standard deviation
```

```python
# seaborn distribution plot
plt.figure(figsize=(15,10))
# figsize tuble of width and height
sns.distplot(data)
plt.show()
```

```python
# 68% rule for normally distributed 
UC = data.mean() + data.std()
LC = data.mean() - data.std()

total = len(data)

# percentage of data between UC and LC
len(data.loc[(data > LC) & (data < UC)])/total * 100
```

```python
# Z transform
z_data = (data - data.mean()) / data.std()
```

```python
# After Z transform Mean = 0
round(z_data.mean())
```

```python
# After Z transform Std = 1
round(z_data.std())
```

```python
# Distribution plot z transformed data
sns.distplot(z_data)
plt.show()
```

### Central Limit Theorm
- All Sample means follow a Normal Distribution
- Mean of Sample Means is approx equal to population mean
- Sample means normal distribution has a standard deviation known as standard error 
    - Standard Error = (Population Standard Deviaton) / squareroot(sample size)

```python
# population -> Follows exponential distribution
Data2 = pd.Series(np.random.exponential(scale=30,size=10000))
```

```python
sns.distplot(Data2)
plt.show()
```

```python
# Tacking samples from population

sample_means = []

for i in range(1,1000):
    s1 = Data2.sample(frac=0.1,replace=False)
    sample_means.append(s1.mean())
    s1 = 0
sample_means = pd.Series(sample_means)
sample_means.head()
```
